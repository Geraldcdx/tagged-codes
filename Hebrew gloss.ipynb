```python
def is_hebrew(text):
    hebrew_range = range(0x0590, 0x05FF)  # Unicode range for Hebrew characters
    return any(ord(char) in hebrew_range for char in text)

def is_english(text):
    english_range = set(range(0x0041, 0x005A + 1))  # Unicode range for English uppercase letters
    english_range |= set(range(0x0061, 0x007A + 1))  # Unicode range for English lowercase letters
    return any(ord(char) in english_range for char in text)

# Test the functions
print(is_hebrew(""))  # True
print(is_hebrew("Hello"))      # False

print(is_english(""))  # False
print(is_english("Hello"))      # True
```

```python
def lan(word):
    if is_hebrew(word):
        return "H"
    else:
        return "E"

def conversion(s):
    indexes = []
    curr = "None"
    prev = "None"
    for index,curr_word in enumerate(s):
        #print(index, "curr_word",curr_word,"prev_word", s[index-1], "indexes", indexes)
        if index == 0:
            indexes.append(index)
        elif lan(curr_word) == "H" and lan(s[index-1]) == "H":#search for previous index and add after for he
            indexes.insert(indexes.index(index-1) + 1, index)
        elif lan(curr_word) == "E" and lan(s[index-1]) == "E": #same for both english
            indexes.insert(indexes.index(index-1) + 1, index)
        elif lan(curr_word) == "H" and lan(s[index-1]) == "E":
            indexes = [index] + indexes
        elif lan(curr_word) == "E" and lan(s[index-1]) == "H":
            indexes = [index] + indexes
        #else:
        #    print(index, curr_word, indexes)
        #    pass
    #print(indexes)
    
    string = ""
    for index in indexes:
        string += s[index] + " "
    return string
```
