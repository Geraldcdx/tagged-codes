```python
#Get data set into the python
genToDeu = pd.read_excel('TOTHT Gen-Deu.xlsx', header = None)
IsaToMal = pd.read_excel('TOTHT Isa-Mal.xlsx', header = None)
JobToSng = pd.read_excel('TOTHT Job-Sng.xlsx', header = None)
JosToEst = pd.read_excel('TOTHT Jos-Est.xlsx', header = None)
torah = np.array(genToDeu)
IsaToMal = np.array(IsaToMal)
JobToSng = np.array(JobToSng)
JosToEst = np.array(JosToEst)
```

```python
myStr = '''Gen,Exo, Lev, Num,Deu, Jos, Jdg, Rut, 1Sa, 2Sa, 1Ki, 2Ki, 1Ch, 2Ch, Ezr, Neh, Est, Job, Psa, Pro, Ecc, Sng,
Isa, Jer, Lam, Ezk, Dan, Hos, Jol, Amo, Oba, Jon, Mic, Nam, Hab, Zep, Hag, Zec, Mal,
Mat, Mrk, Luk, Jhn, Act, Rom, 1Co, 2Co, Gal, Eph, Php, Col, 1Th, 2Th, 1Ti, 2Ti, Tit, Phm, Heb, Jas, 1Pe, 2Pe, 1Jn, 2Jn, 3Jn, Jud, Rev'''
mylist = [x.strip() for x in myStr.split(',')]
mylist.index('Rev') + 1
```

```python
ylt = pd.read_csv('t_ylt-v2.csv')
ylt.head()
```

```python
word = OT
edited = []
for i in range(0,len(word)-1):
    #print(word[i])
    #break
    inflected_word = word[i][2]
    word_component = word[i][5]
    try:
        book = word[i][0].split('.')[0]
        #no_book = mySet[book]
        verse = word[i][0].split('.')[1]
        chapter = word[i][0].split('.')[2].split('-')[0]

        strongs = []
        trans = []
        #print('here',word_component.split('/'), '\n')
        for words in word_component.split('/'):
            try:
                strongs.append(words.split('=')[0])
                trans.append(words.split('=')[2])
            except:
                pass
    except:
        pass
    #print(book, verse, chapter, inflected_word, strongs, '\n')
    edited.append(['1',book,verse, chapter, inflected_word, strongs,trans])
```

FREQ

```python
freq = {}
for word in edited:#edited[0][5])
    for w in word[5]:
        if w not in freq:
            freq[w] =1
        else:
            freq[w] +=1
```

```python
# Example dictionary
#my_dict = {'apple': 5, 'banana': 2, 'orange': 8, 'pear': 1}

# Sort by keys and create a new dictionary
#sorted_dict_keys = dict(sorted(my_dict.items()))

# Sort by values and create a new dictionary
s = dict(sorted(freq.items(), key=lambda x: x[1],reverse = True))

# Print the sorted dictionaries
#print('Sorted by keys:', sorted_dict_keys)
#print('Sorted by values:', sorted_dict_values)
total = np.sum([y for x,y in s.items()])
nums = [y for x,y in s.items()]
```

```python
cum = [0]
for l in range(0, len(nums)):
    cum.append(cum[l] + nums[l])  
```

```python
def getTranslation(book, chapterNo, verseNo):
    temp = ""
    try:
        temp = ylt[(ylt['b'] == mylist.index(book) + 1) 
               & (ylt['c'] == int(chapterNo)) 
               & (ylt['v'] == int(verseNo)) ]['t'].values[0]
    except:
        print(book, chapterNo, verseNo, type(book), type(chapterNo), type(verseNo))
    return temp

#print(getTranslation('Gen', 1, 2))


#ylt[(ylt['b'] == mylist.index('Gen') + 1) & (ylt['c'] == 1) & (ylt['v'] == 2) ]['t'].values[0]
```

```python
verse = ""
strongs = []
new = []
verses = []
trans = []
new_trans = ""
no = 0
for i in range(1,len(edited)-1):
    verse += str(edited[i-1][4]) + ' '
#     print(edited[i-1][4])
    strongs.append(edited[i-1][5])
    trans.append(edited[i-1][6])
    if (edited[i-1][3] != edited[i][3]):
        for row in strongs:
            for item in row:
                new.append(item)
        for row in trans:
            for item in row:
                new_trans+= item + " "
        ylt_verse = getTranslation(edited[i-1][1], edited[i-1][2], edited[i-1][3])
        #print("ylt",ylt_verse)
        #break
        verses.append([edited[i-1][1],  edited[i-1][2],  edited[i-1][3], verse, new, new_trans,ylt_verse])
        #print(getTranslation(edited[i-1][1], int(edited[i-1][2]), int(edited[i-1][3])))
        no = no + 1
        verse = ""
        new = []
        new_trans = ""
        strongs = []
        trans = []
```
